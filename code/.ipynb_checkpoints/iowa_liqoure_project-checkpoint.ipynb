# Data Cleaning

After checking the raw data, I can see that State Bottle Cost, State Bottle Retail, and Sale (Dollars) need to be in float because it contains numerical values. Also, Store Number, County Number, Category, Vendor Number, and Item Number need to be an object.

```python
#Checking the dataframe information. 
df.info()
```

```python
# Set columns State Bottle Cost, State Bottle Retail, Sale (Dollars) 
# data type to float and drop $ sign.
# Converting Store Number column data type to string.
df['State Bottle Cost'] = df['State Bottle Cost'].str.replace('$', '')
df['State Bottle Cost'] = df['State Bottle Cost'].astype(float)

df['State Bottle Retail'] = df['State Bottle Retail'].str.replace('$', '')
df['State Bottle Retail'] = df['State Bottle Retail'].astype(float)

df['Sale (Dollars)'] = df['Sale (Dollars)'].str.replace('$', '')
df['Sale (Dollars)'] = df['Sale (Dollars)'].astype(float)

#I only need Store Number column to be in string.
df['Store Number'] = df['Store Number'].astype(str)
```

```python
df.columns
```

```python
#Changed column names to make it more readable.

new_col = [u'Invoice/Item Number', u'Date', u'Store_Number', u'Store_Name',
           u'Address', u'City', u'Zip_Code', u'Store Location', u'County_Number', 
           u'County', u'Category', u'Category_Name', u'Vendor_Number', u'Vendor Name',
           u'Item_Number', u'Item_Description', u'Pack', u'Bottle_Volume_mL', 
           u'State_Bottle_Cost', u'State_Bottle_Retail', u'Bottles_Sold',
           u'Sale_Dollars', u'Volume_Sold_L', u'Volume_Sold_G']
df.columns = new_col
```

```python
#Re-check the data information to see everything changed correctly.
df.info()
```

```python
#checking how many rows and columns in the dataframe.
df.shape
```

```python
#check and visualize missing values.
df.isnull().sum().plot(kind='bar')
plt.title('Histogram of Null Values')
plt.show();

print(df.isnull().sum())
```

Total missing value counts (28714) are very small compared to total dataset. Therefore, I will drop all missing values and save the changes.

```python
# Drop all null values and save it in the dataframe.
df.dropna(inplace=True)
```

```python
# Checking if missing values got dropped successfully.
df.isnull().sum()
```

```python
#Check/Count total duplicate data.
df.duplicated().sum() 
```

```python
#Drop all duplicate data because its redundant.
df.drop_duplicates(inplace=True)
```

```python
#Change date to datetime.
df['Date'] = pd.to_datetime(df.Date)

#Add columns month and year by sperating both from date.
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
#Set 2016 year records.
year_16 = df[(df.Year == 2016)]

#Set 2015 year records.
year_15 = df[(df.Year == 2015)]
```

```python
#Split 2015-16 into each quarter
q1_15 = year_15[(year_15.Month <= 3)]
q2_15 = year_15[((year_15.Month > 3) & (year_15.Month <= 6))]
q3_15 = year_15[((year_15.Month < 6) & (year_15.Month <=9))]
q4_15 = year_15[(year_15.Month > 9)]
q1_16 = year_16[(year_16.Month <= 3)]
```

```python
# First Quarter Sales for 2015
q1_15_stores = q1_15.groupby('City').Sale_Dollars.mean()

#Second Quarter Sales for 2015
q2_15_stores = q2_15.groupby('City').Sale_Dollars.mean()

#Third Quarter Sales for 2015
q3_15_stores = q3_15.groupby('City').Sale_Dollars.mean()

#Fourth Quarter Sales for 2015
q4_15_stores = q4_15.groupby('City').Sale_Dollars.mean()

#First Quarter Sales for 2016
q1_16_stores = q1_16.groupby('City').Sale_Dollars.mean()
```

```python
#Creating an empty data frame to put all
#quarters sales from 2015 to 2016 1st quarter.
qt_df = pd.DataFrame()

#Creating new columns for all quarters 
#sales from 2015 to 2016 1st quarter.
qt_df['q1_15'] = q1_15_stores
qt_df['q2_15'] = q2_15_stores
qt_df['q3_15'] = q3_15_stores
qt_df['q4_15'] = q4_15_stores
qt_df['q1_16'] = q1_16_stores
qt_df['City'] = qt_df.index

qt_df.head()
```

```python
#Pull sums of all quarterly sales.
qt_df.sum()
```

<a id='Refine the data'></a>
## Refine the data

```python
# Group together by top cities, zip codes, 
# store numbers to see profit and set it to df1.

df1 = df.groupby(["City", "Zip_Code", "Store_Number"], as_index=False)["Profit"]\
.sum().sort_values("Profit", ascending=False).drop_duplicates()

df1.head()
```

```python
#Merge df1(dataframe with City, Zip Code, Store Number, and Profit) 
#with qt_df(all quarterly sales) and set it to df_with_qt.

df_with_qt = pd.merge(df1, qt_df, how='outer', on='City')
```

```python
# Check for all null values
df_with_qt.isnull().sum()
```

Obviously, there are null values because quarterly sales data are broken into each quarter into each column

```python
#Drop all null values in df_with_qt dataframe.
df_with_qt.dropna(inplace=True)
```

To make any statistical relationships, correlations, or other relevant properties of the dataset, I have to make the length of datasets equal to each other and easiest way was to just drop all missing values. The total missing values for this data frame is so small that it's not going to impact my inference.

```python
#Plot heatmap for df_with_qt dataframe to see 
#correlations for all quarters and profits.

def correlation_heat_map(df):
    corrs = df_with_qt.corr()

    # Set the default matplotlib figure size:
    fig, ax = plt.subplots(figsize=(11,7))

    # Generate a mask for the upper triangle (taken from the Seaborn example gallery):
    mask = np.zeros_like(corrs, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Plot the heatmap with Seaborn.
    # Assign the matplotlib axis the function returns. This allow us to resize the labels.
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    ax = sns.heatmap(corrs, mask=mask, annot=True, cmap=cmap, vmin=-1, vmax=1)

    # Resize the labels.
    ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=14, rotation=30)
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=14, rotation=0)

    # If you put plt.show() at the bottom, it prevents those useless printouts from matplotlib.
    plt.title('Profit and Q-sales Relationship Heat Map')
    
df_with_target = df_with_qt
correlation_heat_map(df_with_target)
```

This heat map shows there are weak positive correlations with profit and all quarters. Interestingly, there are strong correlation between 3rd quarter sales and the 1st quarter for 2015, and the 2nd quarter for 2015 out of all quarter sales.

Based on this heat map, I will build a model to test overall 2015 and every 2015 quarters to predict 2016 sales.

<a id='Build models'></a>
# Build models

```python
#Using all 2015 Quarters to predict 2016 Quarter result

y = df_with_qt['q1_16']# Dependent Variable (Y_val)
X = df_with_qt[['q1_15', 'q2_15', 'q3_15', 'q4_15']]# Independent Variables (x_val)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)# Split to train and test sets

scaler = StandardScaler()# Scale the data
scaler.fit(X_train)

X_train = scaler.transform(X_train)# Scale the data on the train/test separately
X_test = scaler.transform(X_test)

lr = LinearRegression()# Perform linear regression
lr.fit(X_train, y_train)

print("Testing All Quarters from 2015")
print("R-squared Score (training set):", lr.score(X_train, y_train))
print("R-squared Score (testing set):", lr.score(X_test, y_test))
print("Intercept:", lr.intercept_)
print("Coefficient:", lr.coef_)
print("Mean of 5-fold CV score:", cross_val_score(lr, X_train, y_train, cv=5).mean())
```

```python
# Testing just first quarter 2015

y = df_with_qt['q1_16']
X = df_with_qt[['q1_15']]


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) # Split to train and test sets

scaler = StandardScaler() # Scale the data 
scaler.fit(X_train)

X_train = scaler.transform(X_train)# Scale the data on the train/test separately
X_test = scaler.transform(X_test)

lr = LinearRegression()# Perform linear regression
lr.fit(X_train, y_train)

print("Testing just 1st Quarters from 2015")
print("R-squared Score (training set):", lr.score(X_train, y_train))
print("R-squared Score (testing set):", lr.score(X_test, y_test))
print("Intercept:", lr.intercept_)
print("Coefficient:", lr.coef_)
print("Mean of 5-fold CV score:", cross_val_score(lr, X_train, y_train, cv=5).mean())
```

```python
#Testing just second quarter 2015

y = df_with_qt['q1_16']
X = df_with_qt[['q2_15']]


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) # Split to train and test sets

scaler = StandardScaler() # Scale the data 
scaler.fit(X_train)

X_train = scaler.transform(X_train)# Scale the data on the train/test separately
X_test = scaler.transform(X_test)

lr = LinearRegression()# Perform linear regression
lr.fit(X_train, y_train)

print("Testing just 2nd quarter 2015")
print("R-squared Score (training set):", lr.score(X_train, y_train))
print("R-squared Score (testing set):", lr.score(X_test, y_test))
print("Intercept:", lr.intercept_)
print("Coefficient:", lr.coef_)
print("Mean of 5-fold CV score:", cross_val_score(lr, X_train, y_train, cv=5).mean())
```

```python
#Testing just third quarter 2015

y = df_with_qt['q1_16']
X = df_with_qt[['q3_15']]


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) # Split to train and test sets

scaler = StandardScaler() # Scale the data 
scaler.fit(X_train)

X_train = scaler.transform(X_train)# Scale the data on the train/test separately
X_test = scaler.transform(X_test)

lr = LinearRegression()# Perform linear regression
lr.fit(X_train, y_train)

print("Testing just 3rd quarter 2015")
print("R-squared Score (training set):", lr.score(X_train, y_train))
print("R-squared Score (testing set):", lr.score(X_test, y_test))
print("Intercept:", lr.intercept_)
print("Coefficient:", lr.coef_)
print("Mean of 5-fold CV score:", cross_val_score(lr, X_train, y_train, cv=5).mean())
```

```python
#Testing just fourth quarter 2015

y = df_with_qt['q1_16']
X = df_with_qt[['q4_15']]


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) # Split to train and test sets

scaler = StandardScaler() # Scale the data 
scaler.fit(X_train)

X_train = scaler.transform(X_train)# Scale the data on the train/test separately
X_test = scaler.transform(X_test)

lr = LinearRegression()# Perform linear regression
lr.fit(X_train, y_train)

print("Testing just 4th quarter 2015")
print("R-squared Score (training set):", lr.score(X_train, y_train))
print("R-squared Score (testing set):", lr.score(X_test, y_test))
print("Intercept:", lr.intercept_)
print("Coefficient:", lr.coef_)
print("Mean of 5-fold CV score:", cross_val_score(lr, X_train, y_train, cv=5).mean())
```
